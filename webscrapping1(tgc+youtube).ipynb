```python
import bs4
import requests
```

```python
url='https://www.tgcindia.com/blog/'
data=requests.get(url)
soup=bs4.BeautifulSoup(data.text,'html.parser')
print(soup.prettify())
```

```python
for para in soup.find('p'):
    print(para)
#will only print the first paragraph and will print nothing if the first paragraph is empty
```

```python
for para in soup.find_all('p'):
    print(para)
```

```python
for para in soup.find_all('p'):
    print(para.text)
```

```python
f=open("tgcpara.txt","w+")
```

```python
f=open("tgcpara.txt","a+")
for para in soup.find_all('p'):
    f.write(para.text)
```

```python
for para in soup.find_all('h3'):
    print(para)
```

```python
for para in soup.find_all('h3'):
    print(para.text)
```

```python
for link in soup.find_all('a'):
    link=link.get('href')
    if link!='#' and link!=None and link!='#login':
        print(link)
```

# webscapping of youtube links

```python
url2='https://www.youtube.com/results?search_query=tgcindia'
data2=requests.get(url2)
soup=bs4.BeautifulSoup(data2.text,'html.parser')
```

```python
for para in soup.find_all('a'):
    link=para.get('href')
    if link[0:3]=="/wa":
        print("https://www.youtube.com"+link)
```
