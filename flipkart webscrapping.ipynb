```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url='https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_0_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_0_2_na_na_na&as-pos=0&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=db3c569d-0c70-40ee-9076-be6f3cf7ea52&as-backfill=on'
```

```python
urlreq=urlopen(url)
data_html=urlreq.read()
```

```python
data_html
```

```python
data_soup=BeautifulSoup(data_html,'html.parser')
print(data_soup.prettify())
```

```python
laptops=data_soup.find_all('div',attrs={'class':'_3O0U0u'})
```

```python
print(laptops)
```

```python
print(len(laptops))
```

```python
laptop=laptops[0]
laptop
```

```python
name=laptop.find_all('div',attrs={'class':'_3wU53n'})
name=name[0]
name=name.text
name=name.split('-')
final_name=name[0]
print(final_name)
```

```python
rating=laptop.find_all('div',attrs={'class':'niH0FQ'})
rating=rating[0].text
print(rating)
```

```python
price=laptop.find_all('div',attrs={'class':'_1vC4OE _2rQ-NK'})
price=price[0].text
print(price)
price=price.replace('₹','')
price=price.replace(',','')
print(price)
```

```python
LAPTOPS=[['Name','Rating','Price']]
for laptop in laptops:
    name=laptop.find_all('div',attrs={'class':'_3wU53n'})
    name=name[0]
    name=name.text
    name=name.split('-')
    final_name=name[0]
    
    rating=laptop.find_all('div',attrs={'class':'niH0FQ'})
    rating=rating[0].text
    
    price=laptop.find_all('div',attrs={'class':'_1vC4OE _2rQ-NK'})
    price=price[0].text
    price=price.replace('₹','')
    price=price.replace(',','')
    
    LAPTOPS.append([final_name,rating,price])
```

```python
print(LAPTOPS)
```
