```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url='https://www.flipkart.com/search?q=samsung%20mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
urlreq=urlopen(url)
data_html=urlreq.read()
```

```python
data_soup=BeautifulSoup(data_html,'html.parser')
print(data_soup.prettify())
```

```python
phones=data_soup.find_all('div',attrs={'class':'_3O0U0u'})
```

```python
PHONES=[['Name','Rating','Price']]
text_file=open('mobiles.txt','w+')
for phone in phones:
    name=phone.find_all('div',attrs={'class':'_3wU53n'})
    name=name[0]
    name=name.text
    name=name.split('-')
    final_name=name[0]
    text_file.write(final_name)
    text_file.write('\t')
    rating=phone.find_all('div',attrs={'class':'niH0FQ'})
    rating=rating[0].text
    text_file.write(rating)
    text_file.write('\t')
    price=phone.find_all('div',attrs={'class':'_1vC4OE _2rQ-NK'})
    price=price[0].text
    price=price.replace('₹','')
    price=price.replace(',','')
    text_file.write(price)
    text_file.write('\n')
    PHONES.append([final_name,rating,price])
text_file.close()
```

```python
from tabulate import tabulate
```

```python
print(tabulate(PHONES,tablefmt='fancy_grid'))
```
